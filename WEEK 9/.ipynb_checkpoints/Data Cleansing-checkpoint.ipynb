```python
def nullVal(bank_df):
    missing_num= bank_df.isnull().sum()
    missing_percent = (100 * missing_num/ len(bank_df))
    concatinated_value = pd.concat([missing_num, missing_percent], axis = 1)
    
    missing = concatinated_value.rename(columns = {0:"Missing ", 1:"Percent of Missing Values"})
    in_sorted_order = missing[missing.iloc[:,1]!=0].sort_values(by= 'Percent of Missing Values', ascending = False).round(1)
    print("Full Bank Data has a total of '{}' coloumns from which the '{}' coloumns are having missing values.".format(bank_df.shape[1],in_sorted_order.shape[0]))
    return in_sorted_order
```

```python
missing_num = nullVal(bank_df)
```

# Imputation using Median
From bank_df.describe(), median is 50% value which is 38

```python
cols = ['age','campaign','pdays','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
plt.figure(figsize=(10,15))
for i, col in enumerate(cols):
    plt.subplot(4,3,i+1)
    bank_df.boxplot(col)
    plt.grid()
    plt.tight_layout()
```

```python
column= ['age']
plt.figure(figsize=(22,5))
n = 1
for col in column:
    ax= plt.subplot(1,2,n)
    sns.boxplot(x=bank_df[col], color='#8FBC8F')
    plt.xlabel(col)
    n +=1
    plt.show()
```

Most of the numerical columns are not normally distributed and some of them have outliers. We can see from the above boxplot that there are outliers available in certain variables.
